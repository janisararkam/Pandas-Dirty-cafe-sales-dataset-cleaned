```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce').astype('Int64') #Converting the data type of the columns into another data type of the data set 
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], errors="coerce")  #converting the transaction date column of the data set into data and time format  
```

```python
df.dtypes #So from here all the columns and its data type are correct of the data set
```

```python
df.isnull().sum() #finding the null values of the columns of the data set 
```

```python
#Converting the UNKNOWN and ERROR value also to  NaN ,so that data set would be proper clean 
df.replace(["UNKNOWN", "ERROR"], pd.NA, inplace=True)
```

```python
#fill the null value of the columns of the data set
df['Item'] = df['Item'].fillna(df['Item'].mode()[0])
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].median())

# Total Spent re-calculate
df['Total Spent'] = df['Quantity'] * df['Price Per Unit']

df['Payment Method'] = df['Payment Method'].fillna(df['Payment Method'].mode()[0])
df['Location'] = df['Location'].fillna(df['Location'].mode()[0])

# Date column fill
df['Transaction Date'] = df['Transaction Date'].ffill()
```

```python
df.isnull().sum() #So from here there is No null value present in the columns of the data set
```

```python
df.rename(columns={'Transaction ID':'Transaction_ID'},inplace=True)  #Renaming the columns of the data set
df.rename(columns={'Item':'Items'},inplace=True)
df.rename(columns={'Quantity':'Quantities'},inplace=True)
df.rename(columns={'Price Per Unit':'Price_Per_Unit'},inplace=True)
df.rename(columns={'Total Spent':'Total_Spent'},inplace=True)
df.rename(columns={'Payment Method':'Payment_Method'},inplace=True)
df.rename(columns={'Location':'Locations'},inplace=True)
df.rename(columns={'Transaction Date':'Transaction_Date'},inplace=True)
```
